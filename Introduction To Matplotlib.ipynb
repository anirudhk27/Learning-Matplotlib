## PLotting from Pandas DataFrame

```python
import pandas as pd
```

```python
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$\,\.]','', regex=True)
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"] =pd.date_range("1/1/2020",periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total"]=car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total");
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
car_sales.plot.bar()
```

```python
car_sales["Odometer (KM)"].plot.hist();
```

```python
car_sales.plot.hist();
```

```python
heart_diseases=pd.read_csv("heart-disease.csv")
```

```python
heart_diseases
```

```python

over_50=heart_diseases[heart_diseases["age"]>50]
```

```python
over_50
```

```python
over_50.plot(x="age",y="chol",kind="scatter",c="target");
```

```python
fig,ax=plt.subplots()
x=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart Dieases and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol Levels")

ax.legend(*x.legend_elements(),title="Target")
ax.axhline(250,linestyle="solid");
```
